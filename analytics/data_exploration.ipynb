```python
#Check the CompetitionDistance mean based on StoreType
median_by_store_type = df_store.groupby(['StoreType'])['CompetitionDistance'].median()
print("Median Competition Distance by Store Type:\n", median_by_store_type)
```

```python
#Check the CompetitionDistance mean based on AssortmentType
median_by_assortment = df_store.groupby('Assortment')['CompetitionDistance'].mean()
print('Median Competition Distance by Assortment Type:\n', median_by_assortment)
```

My realization:
First, the store type is divided into 4 categories: a,b,c,d but the document description does not provide further information about what are the differences between those categories. Therefore, it is not reasonable to impute null value based on StoreType.

Second, the assortment type is divided into 3 categories: a = basic, b = extra, c = extended. I then assume that the competition distance will decrease from category a to c (The more complicated the store, the more competition will appear). But the data shows that:
Median Competition Distance by Assortment Type:
Assortment
a    4400.338409
b    1177.777778
c    6638.769531
Which means that it does not decrease along with the assortment type, and it does not match my assumption. So I think it's also unreasonable to impute the null value based on assortment type.

Therefore, we will use the overall mean/median to impute these 3 null values because it the least risky way. 

```python
#Replace null value with the median value because its 2 times lower than the mean value
val = df_store['CompetitionDistance'].median()
df_store['CompetitionDistance'].fillna(val, inplace=True)
```

```python
#Check again the number of null values of CompetitionDistance variable
df_store['CompetitionDistance'].isnull().sum()
```

Merge the 2 datasets for more information

```python
#Join 2 datasets together
train_store = pd.merge(df_train, df_store, how='left', on='Store')
train_store
```

We will generate heatmap to see the correlation between Sales (target variable) and other predictive variables

```python
corr_matrix = train_store.drop('Open', axis=1).corr(numeric_only = True)
corr_matrix
```

```python
# Generate a mask for the upper triangle
mask = np.zeros_like(corr_matrix, dtype = bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize = (11, 9))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_matrix, mask = mask,
            square = True, linewidths = .5, ax = ax, cmap = "BuPu")      
plt.show()
```

We can see that there is a high correlation between Sales and number of Customers.
Promo is also a variable that has positive impact on Sales
Promo2 even has negative impacts on Sales

We will start looking closely at each variable and its impact on Sales

First we will look at the sales descriptive data based on store type

As we can see, Store Type B has the highest sales volumns, while the others demonstrate similar sales level. What about Sales per customer?

Interestingly, although store type b has the highest volumns of sales, it is the one that has lowest sales per customer. It demonstrates that store type b sells goods in the low price range, while the opposite is true for store type d.

We will have a graph to see the patterns of sales, customers, and sales per customer

```python
# Create relplot for Sales by DayOfWeek and StoreType
plt.figure(figsize=(15, 10))
sns.relplot(data=train_store, 
            x='Month', 
            y='Sales',
            col='DayOfWeek',
            row='StoreType',
            hue='StoreType',
            kind='line',
            height=3,
            aspect=1.5)

plt.suptitle('Sales Trends by Store Type and Day of Week', y=1.02)
plt.tight_layout()
plt.show()

# Create relplot for Customers by DayOfWeek and StoreType
plt.figure(figsize=(15, 10))
sns.relplot(data=train_store, 
            x='Month', 
            y='Customers',
            col='DayOfWeek',
            row='StoreType',
            hue='StoreType',
            kind='line',
            height=3,
            aspect=1.5)

plt.suptitle('Customer Trends by Store Type and Day of Week', y=1.02)
plt.tight_layout()
plt.show()

# Create relplot for Sales per Customer by DayOfWeek and StoreType
plt.figure(figsize=(15, 10))
sns.relplot(data=train_store, 
            x='Month', 
            y='Sales Per Customer',
            col='DayOfWeek',
            row='StoreType',
            hue='StoreType',
            kind='line',
            height=3,
            aspect=1.5)

plt.suptitle('Sales per Customer Trends by Store Type and Day of Week', y=1.02)
plt.tight_layout()
plt.show()
```
